```python
def setZeroes(matrix):
    row = len(matrix)
    column = len(matrix[0])
    I = []
    J = []
    for i in range(row):
        for j in range(column):
            if matrix[i][j] == 0:
                I.append(i)
                J.append(j)
    for i in range(row):
        for j in range(column):
            if i in I or j in J:
                matrix[i][j] = 0
    return matrix

setZeroes([[1, 2, 3, 4],
           [5, 0, 7, 8],
           [0,10,11,12],
           [13,14,15,0]])
```
