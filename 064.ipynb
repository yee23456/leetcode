```python
def minPathSum(grid):
    #先確定長寬
    m = len(grid) #行
    n = len(grid[0]) #列
    for i in range(1,n):
        grid[0][i]+=grid[0][i-1]
    
    for i in range(1,m):
        grid[i][0]+= grid[i-1][0]
    
    for i in range(1,m):#列
        for j in range(1,n):#行
            grid[i][j]+= min(grid[i-1][j],grid[i][j-1])
    return grid[-1][-1]
minPathSum([[1,2,3],[4,5,6]])
    
```
