# 別人寫的

```python
class Solution:
    def singleNumber(self, nums: List[int]) -> int:
        if len(nums) == 1:  # always think about base cases: he we have one
            return nums[0]
        nums.sort()  # sorting will take O(nlogn)
        last_element = None  # set last seen as None
        count = 1  # this is our counter of numbers
        for el in (nums + [0]):  # we are going through sequence, [0] triggers last change of numbers
            if last_element is None:  # check whether last_element is None or not - don't write "not ..." cause if last_element is 0, it will ruin the solution!
                last_element = el
                continue
            elif el == last_element:  # if we have seen an element - we increase counter by 1
                count += 1
            else:
                if count == 1:  # if not - we check counter 
                    return last_element
                count = 1  # and return it to the very first state
                last_element = el  # assign el to variable last_element
```
