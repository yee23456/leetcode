```python
def combinationSum(target):
    answer = []
    candidates = [2,3,5]
    universe = [([], 0)]
    for n in candidates:
        for (ls, v) in universe:
            if v + n == target:
                answer.append(ls + [n])
            elif v + n < target:
                print(ls)
                universe.append((ls + [n], v + n))
    return #print(answer)
combinationSum(7)
```

```python
def combinationSum3(self, k: int, n: int) -> List[List[int]]:
    result = []
    self._helperCombo3(1, k, n, [], result)
    return result

def _helperCombo3(self, curr_num, k, n, ans, result):
    if k < 0 or n < 0:
        return
    if k == 0 and n == 0:
        res = ans.copy()
        result.append(res)
        return
    for i in range(curr_num, 10):
        ans.append(i)
        self._helperCombo3(i + 1, k - 1, n - i, ans, result)
        ans.pop()
```
