```python
#我想到的
def rotate(matrix):
    matrix.reverse()
    for i in range(len(matrix)):
        for j in range(i):
            matrix[i][j] ,matrix[j][i] = matrix[j][i] ,matrix[i][j]
    return matrix
            
            
            
rotate([[1,2,3],[4,5,6],[7,8,9]])


```

```python
#別人想到的
class Solution:
    def rotate(self, A):
        n = len(A)
        for i in range(n/2):
            for j in range(n-n/2):
                A[i][j], A[~j][i], A[~i][~j], A[j][~i] = \
                         A[~j][i], A[~i][~j], A[j][~i], A[i][j]
```
